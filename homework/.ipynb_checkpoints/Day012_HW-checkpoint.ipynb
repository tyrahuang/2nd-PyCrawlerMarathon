### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
#print(r.text)
soup = BeautifulSoup(r.text, "html5lib")


#x = soup.find(class_="part_list_2").find_all('h3')
#print(x)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find('em').text,d.find_all('a')[-1].text)
    
    #print(d.find(class_="date").text)
    #print( d.find_all('a'))
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
### dict{key,value 用List}
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
#print(r.text)
soup = BeautifulSoup(r.text, "html5lib")

#x = soup.find(class_="part_list_2").find_all('h3')
#print(x)
category = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    category.setdefault(d.find('em').text,[]).append(d.find_all('a')[-1].text)
   
for i in category.items():
    print (i)
```

### ③ 哪一個類別的文章最多

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
#print(r.text)
soup = BeautifulSoup(r.text, "html5lib")

#x = soup.find(class_="part_list_2").find_all('h3')
#print(x)
category = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    category.setdefault(d.find('em').text,[]).append(d.find_all('a')[-1].text)

for i in category.items():
    print(i[0], len(i[1]))
```
